# Modeling non-linear data using Python's numpy library

## 1st example (U-shaped relationship)

```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate random x values
x = np.linspace(-5, 5, 100)

# Generate corresponding y values with a U-shaped relationship
np.random.seed(42)
y = 2 * x**2 - 3 * x + 1 + np.random.normal(0, 5, size=len(x))

# Create scatter plot of the data
plt.scatter(x, y)

# Compute polynomial fit of 2nd order
fit = np.polyfit(x, y, 2)
p = np.poly1d(fit)

# Get coefficients
coeffs = p.coef
print("Coefficients:", coeffs)

# Calculate the R-squared value
y_mean = np.mean(y)
ss_res = np.sum((y - p(x))**2)
ss_tot = np.sum((y - y_mean)**2)
r_squared = 1 - (ss_res / ss_tot)
print("R-squared:", round(r_squared, 4))

# Create line showing the polynomial fit
plt.plot(x, p(x), 'r')

# Set axis labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('U-shaped relationship')
plt.grid()

# Display plot
plt.show()
```

## 2nd example (inverse U-shaped relationship)

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random x values
x = np.linspace(-5, 5, 100)

# Generate corresponding y values with an inverse U-shaped relationship
np.random.seed(42)
y = -4 * x**2 + 3 * np.abs(x) + np.random.normal(0, 5, size=len(x))

# Create scatter plot of the data
plt.scatter(x, y)

# Compute polynomial fit of 2nd order
fit = np.polyfit(x, y, 2)
p = np.poly1d(fit)

# Get coefficients
coeffs = p.coef
print("Coefficients:", coeffs)

# Calculate the R-squared value
y_mean = np.mean(y)
ss_res = np.sum((y - p(x))**2)
ss_tot = np.sum((y - y_mean)**2)
r_squared = 1 - (ss_res / ss_tot)
print("R-squared:", round(r_squared, 4))

# Create line showing the polynomial fit
plt.plot(x, p(x), 'r')

# Set axis labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Inverse U-shaped relationship')
plt.grid()

# Display plot
plt.show()
```
